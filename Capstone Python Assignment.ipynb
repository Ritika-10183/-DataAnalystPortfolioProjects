```python
# Task 2

import pandas as pd
import numpy as np

# Original Project DataFrame (with some missing values introduced for demonstration)
project_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005', 'A002', 'A005', 'A003', 'A001', 'A003', 'A001', 'A004', 'A004', 'A005'],
    'Project': ['Project 1', 'Project 2', 'Project 3', 'Project 4', 'Project 5', 'Project 6', 'Project 7', 'Project 8', 
                'Project 9', 'Project 10', 'Project 11', 'Project 12', 'Project 13', 'Project 14'],
    'Cost': [1002000, 2000000, 4500000, np.nan, 1000000, 680000, 400000, 350000, np.nan, 300000, 2000000, np.nan, 3000000, 200000],
    'Status': ['Finished', 'Ongoing', 'Finished', 'Ongoing', 'Finished', 'Failed', 'Finished', 'Failed', 'Ongoing', 
               'Finished', 'Failed', 'Ongoing', 'Finished', 'Finished']
}
project_df = pd.DataFrame(project_data)

# Replace missing values in 'Cost' with running average using a 'for' loop
total_cost = 0
count = 0

for i in range(len(project_df)):
    if np.isnan(project_df.loc[i, 'Cost']):
        if count > 0:
            project_df.loc[i, 'Cost'] = total_cost / count
    else:
        # Update the running sum and count
        total_cost += project_df.loc[i, 'Cost']
        count += 1

# Display the modified DataFrame
print("Modified Project DataFrame with missing 'Cost' values replaced by running average:")
print(project_df)
```

```python
# Task 3

import pandas as pd

# Original Employee DataFrame
employee_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],
    'Name': ['John Alter', 'Alice Luxumberg', 'Tom Sabestine', 'Nina Adgra', 'Amy Johny'],
    'Gender': ['M', 'F', 'M', 'F', 'F'],
    'City': ['Paris', 'London', 'Berlin', 'Newyork', 'Madrid'],
    'Age': [25, 27, 29, 31, 30]
}
employee_df = pd.DataFrame(employee_data)

# Split the 'Name' column into 'First Name' and 'Last Name'
employee_df[['First Name', 'Last Name']] = employee_df['Name'].str.split(' ', expand=True)

# Remove the original 'Name' column
employee_df = employee_df.drop(columns=['Name'])

# Display the modified DataFrame
print("Modified Employee DataFrame with 'First Name' and 'Last Name' columns:")
print(employee_df)
```

```python
# Task 4

import pandas as pd
import numpy as np

# Project DataFrame
project_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005', 'A002', 'A005', 'A003', 'A001', 'A003', 'A001', 'A004', 'A004', 'A005'],
    'Project': ['Project 1', 'Project 2', 'Project 3', 'Project 4', 'Project 5', 'Project 6', 'Project 7', 'Project 8', 
                'Project 9', 'Project 10', 'Project 11', 'Project 12', 'Project 13', 'Project 14'],
    'Cost': [1002000, 2000000, 4500000, np.nan, 1000000, 680000, 400000, 350000, np.nan, 300000, 2000000, np.nan, 3000000, 200000],
    'Status': ['Finished', 'Ongoing', 'Finished', 'Ongoing', 'Finished', 'Failed', 'Finished', 'Failed', 'Ongoing', 
               'Finished', 'Failed', 'Ongoing', 'Finished', 'Finished']
}
project_df = pd.DataFrame(project_data)

# Employee DataFrame
employee_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],
    'First Name': ['John', 'Alice', 'Tom', 'Nina', 'Amy'],
    'Last Name': ['Alter', 'Luxumberg', 'Sabestine', 'Adgra', 'Johny'],
    'Gender': ['M', 'F', 'M', 'F', 'F'],
    'City': ['Paris', 'London', 'Berlin', 'Newyork', 'Madrid'],
    'Age': [25, 27, 29, 31, 30]
}
employee_df = pd.DataFrame(employee_data)

# Seniority Level DataFrame
seniority_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],
    'Designation Level': [2, 2, 3, 2, 3]
}
seniority_df = pd.DataFrame(seniority_data)

# Merge the dataframes
# Step 1: Merge Project DataFrame with Employee DataFrame on 'ID'
project_employee_df = pd.merge(project_df, employee_df, on='ID', how='left')

# Step 2: Merge the result with the Seniority Level DataFrame on 'ID'
final_df = pd.merge(project_employee_df, seniority_df, on='ID', how='left')

# Display the final combined DataFrame
print("Final DataFrame:")
print(final_df)
```

```python
# Task 5

# Add a 'Bonus' column based on project completion status and cost
final_df['Bonus'] = final_df.apply(lambda row: row['Cost'] * 0.05 if row['Status'] == 'Finished' else 0, axis=1)

# Display the updated Final DataFrame
print("Final DataFrame with Bonus Column:")
print(final_df)
```

```python
# Task 6

# Step 1: Demote designation level by 1 for employees with failed projects
final_df['Designation Level'] = final_df.apply(
    lambda row: row['Designation Level'] - 1 if row['Status'] == 'Failed' else row['Designation Level'],
    axis=1
)

# Step 2: Delete records where the designation level is above 4
final_df = final_df[final_df['Designation Level'] <= 4].reset_index(drop=True)

# Display the updated Final DataFrame
print("Final DataFrame after demotion and deletion:")
print(final_df)
```

```python
# Task 7

# Add "Mr." or "Mrs." to the First Name based on Gender
final_df['First Name'] = final_df.apply(
    lambda row: f"Mr. {row['First Name']}" if row['Gender'] == 'M' else f"Mrs. {row['First Name']}", 
    axis=1
)

# Drop the 'Gender' column
final_df = final_df.drop(columns=['Gender'])

# Display the modified Final DataFrame
print("Final DataFrame with titles added to 'First Name' and 'Gender' column removed:")
print(final_df)
```

```python
# Task 8

# Promote designation level by 1 for employees older than 29
final_df['Designation Level'] = final_df.apply(
    lambda row: row['Designation Level'] + 1 if row['Age'] > 29 else row['Designation Level'],
    axis=1
)

# Display the modified Final DataFrame
print("Final DataFrame with updated Designation Level based on age:")
print(final_df)
```

```python
# Task 9

# Calculate total project cost for each employee
total_proj_cost_data = final_df.groupby(['ID', 'First Name'])['Cost'].sum().reset_index()

# Rename the 'Cost' column to 'Total Cost' for clarity
total_proj_cost_data = total_proj_cost_data.rename(columns={'Cost': 'Total Cost'})

# Display the new TotalProjCost DataFrame
print("TotalProjCost DataFrame:")
print(total_proj_cost_data)
```

```python
# Task 10

# Filter employees whose city name contains the letter "o"
employees_with_o_in_city = final_df[final_df['City'].str.contains('o', case=False, na=False)]

# Display the filtered DataFrame
print("Employees whose City contains the letter 'o':")
print(employees_with_o_in_city)
```
